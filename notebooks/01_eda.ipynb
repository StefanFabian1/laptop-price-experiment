# 01 – EDA

```python
import os
import shutil
import pandas as pd

from src.cleaning import build_combined_dataset, collect_clean_datasets

raw_dir = '../data/raw'
os.makedirs(raw_dir, exist_ok=True)
files = [f for f in os.listdir(raw_dir) if f.endswith('.csv')]

if not files:
    fallback_dir = '../src/raw'
    fallback_files = [f for f in os.listdir(fallback_dir) if f.endswith('.csv')]
    assert fallback_files, 'Vlož Kaggle CSV do data/raw/ alebo src/raw/'
    src_path = os.path.join(fallback_dir, fallback_files[0])
    dst_path = os.path.join(raw_dir, fallback_files[0])
    shutil.copy2(src_path, dst_path)
    files = [fallback_files[0]]

path = os.path.join(raw_dir, files[0])
print(f'Načítavam: {os.path.basename(path)}')
df_raw = pd.read_csv(path)
df_raw.head()
```

# Základné čistenie a uloženie

```python
datasets = collect_clean_datasets(df_raw)
for name, subset in datasets.items():
    print(f'{name}: {len(subset)} riadkov')

df_clean = pd.concat(datasets.values(), ignore_index=True)
print('Combined shape:', df_clean.shape)
print(df_clean.describe(include='all'))

os.makedirs('../data/processed', exist_ok=True)
df_clean.to_csv('../data/processed/clean.csv', index=False)
print('Uložené do data/processed/clean.csv')
```
